```python
import csv
import pandas as pd


laptop = pd.read_csv('C:/Users/user/OneDrive/Desktop/laptops.CSV',decimal = ',' )

laptop.columns = ['brand', 'model', 'category', 'screen_size', 'screen', 'cpu', 'ram',
                 'storage', 'gpu', 'op_system', 'op_system_version', 'weight', 'price']
laptop
```

```python
#1.----N ամենաթանկ laptop-ները f’{brand} {model} {price}’ այս ֆորմատով

def expensive_laptop(n):
    return laptop.nlargest(n, 'price')[['brand','model','price']]

n = int(input("Input count:"))
print(f'The most expensive {n} laptops.')
expensive_laptop(n)
```

```python
#2.----N ամենամատչելի laptop-ները f’{brand} {model} {price}’ այս ֆորմատով

def cheapest_laptop(n):
    return laptop.nsmallest(n, 'price')[['brand','model','price']]

n = int(input("Input count:"))
print(f'The cheapest {n} laptops.')
cheapest_laptop(n)
```

```python
#3.----բոլոր օպերացիոն համակարգերը և իրենց laptop-երի քանակը(օրինակ՝ windows: 456, macOS:46, Linux: 876...)
count_laptop_op_system = laptop['op_system'].value_counts()
count_laptop_op_system
```

```python
#4.----N ամենածանր laptop-ները f’{brand} {model} {weight}’ այս ֆորմատով

def heavy_laptop(n):
    try:
        laptop['weights'] = laptop['weight'].apply(lambda w: float(w.strip('kg').strip('kgs')))
        return laptop.nlargest(n, 'weights')[['brand','model','weight']]
    except ValueError:
        print("Oops! Something went wrong.")
        print("We only have used kg(kgs) to measure the weight of laptops")
n = int(input("Input count:"))
heavy_laptop(n)
```

```python
#5.----N ամենահզոր RAM ունեցող laptop-ները f’{brand} {model} {ram}’ այս ֆորմատով

def powerfull_ram(n):
    try:
        laptop['rams'] = laptop.ram.apply(lambda r: int(r.strip('GB')))
        return laptop.nlargest(n, 'rams')[['brand','model','ram']]
    except ValueError:
        print("Oops! Something went wrong.")
        print("We only have used GB to measure the power of laptop's RAM")
    
n = int(input("Input count:"))
powerfull_ram(n)
```

```python
#6.----բոլոր չափի RAM-երը և իրենց մոդելների քանակը(օրինակ՝ 4GB: 456, 8GB:46, 10GB: 876...)
count_laptop_ram = laptop['ram'].value_counts()
count_laptop_ram
```

```python
#7.----էկրանի չափսերը և իրենց մոդելների քանակը։ 
#էկրանի չափսերը խմբավորել հետևյալ կերպ։ մինչև 10”, 10”-13”, 13”-15” և 15” կամ ավելի մեծ
size = laptop['screen_size'].apply(lambda size:float(size[:-1]))

a,b,c,d = 0,0,0,0

for i in range(len(laptop.index)):
    current_size = size[i]
    if current_size < 10:
        a += 1
    elif current_size >= 10 and current_size < 13:
        b += 1
    elif current_size >= 13 and current_size < 15:
        c += 1
    else:
        d += 1
print(
    f'''until 10”: {a},
10”-13”: {b},
13”-15”: {c},
15” or more  {d} '''
)
```

```python
#8.----բոլոր brand-երը և իրենց մոդելների քանակը(օրինակ՝ apple: 11, Dell:80...)
count_laptop_brand = laptop['brand'].value_counts()
count_laptop_brand
```
